A distribution is discrete if it is associated with a discrete random variable, which takes on countable values (finite or countably infinite).

```python
#Q(2)Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def generate_discrete_uniform(a, b, size=10000):
    """Generate samples from a discrete uniform distribution."""
    return np.random.randint(a, b + 1, size)

def plot_pmf(a, b, samples):
    """Plot the Probability Mass Function (PMF) for a discrete uniform distribution."""
    values, counts = np.unique(samples, return_counts=True)
    pmf = counts / sum(counts)

    plt.bar(values, pmf, color='lightcoral', edgecolor='black', alpha=0.7)
    plt.xlabel('Random Variable Value')
    plt.ylabel('Probability Mass Function (PMF)')
    plt.title(f'Discrete Uniform Distribution U({a},{b})')
    plt.xticks(range(a, b+1))
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Parameters for the discrete uniform distribution
a, b = 1, 10  # Change the range as needed
samples = generate_discrete_uniform(a, b)
plot_pmf(a, b, samples)
```

```python
#Q(3)Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

def bernoulli_pdf(p):
    """Calculate and plot the probability distribution function (PDF) of a Bernoulli distribution."""
    x = [0, 1]  # Possible outcomes in a Bernoulli trial
    pmf = [1 - p, p]  # Probability mass function

    plt.bar(x, pmf, color='blue', edgecolor='black', alpha=0.7)
    plt.xticks([0, 1])
    plt.xlabel('Outcome')
    plt.ylabel('Probability')
    plt.title(f'Bernoulli Distribution PDF (p={p})')
    plt.ylim(0, 1)
    plt.show()

    return dict(zip(x, pmf))

# Example usage
p = 0.6  # Probability of success
bernoulli_pdf(p)
```

```python
#(Q)6Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_cdf(a, b, size=10000):
    """Generate samples from a discrete uniform distribution and compute the CDF."""
    samples = np.random.randint(a, b + 1, size)
    values, counts = np.unique(samples, return_counts=True)
    cdf = np.cumsum(counts) / sum(counts)

    # Plot CDF
    plt.step(values, cdf, where='mid', color='red', label='Empirical CDF')
    plt.xlabel('Random Variable Value')
    plt.ylabel('Cumulative Probability')
    plt.title(f'Cumulative Distribution Function (CDF) of U({a},{b})')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.show()

    return values, cdf

# Parameters
a, b = 1, 6  # Example: Rolling a fair 6-sided die
calculate_cdf(a, b)
```

```python
#Q(9)Write a Python function to calculate Z-scores from a dataset and plot them?
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def simulate_normal(mu, sigma, size=10000):
    """Generate samples from a normal distribution and plot the histogram."""
    samples = np.random.normal(mu, sigma, size)

    # Plot histogram
    plt.hist(samples, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

    # Plot theoretical normal distribution curve
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    pdf = stats.norm.pdf(x, mu, sigma)
    plt.plot(x, pdf, 'r-', lw=2, label='Normal PDF')

    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.title(f'Normal Distribution (Î¼={mu}, Ïƒ={sigma})')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    return samples

def calculate_z_scores(data):
    """Calculate Z-scores from a dataset and plot them."""
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev

    # Plot Z-scores histogram
    plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='green', edgecolor='black')
    plt.xlabel('Z-Score')
    plt.ylabel('Density')
    plt.title('Z-Score Distribution')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    return z_scores

# Parameters
mu, sigma = 0, 1  # Mean and standard deviation
data = simulate_normal(mu, sigma)
calculate_z_scores(data)
```

```python
#Q(16)Create a confidence interval for a dataset using Python and interpret the rESULT?
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def simulate_normal(mu, sigma, size=10000):
    """Generate samples from a normal distribution and plot the histogram."""
    samples = np.random.normal(mu, sigma, size)

    # Plot histogram
    plt.hist(samples, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

    # Plot theoretical normal distribution curve
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    pdf = stats.norm.pdf(x, mu, sigma)
    plt.plot(x, pdf, 'r-', lw=2, label='Normal PDF')

    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.title(f'Normal Distribution (Î¼={mu}, Ïƒ={sigma})')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    return samples

def calculate_z_scores(data):
    """Calculate Z-scores from a dataset and plot them."""
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev

    # Plot Z-scores histogram
    plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='green', edgecolor='black')
    plt.xlabel('Z-Score')
    plt.ylabel('Density')
    plt.title('Z-Score Distribution')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    return z_scores

def calculate_z_score(data_point, data):
    """Calculate the Z-score for a given data point and compare it to a standard normal distribution."""
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (data_point - mean) / std_dev

    # Plot standard normal distribution
    x = np.linspace(-4, 4, 1000)
    pdf = stats.norm.pdf(x, 0, 1)

    plt.plot(x, pdf, 'b-', lw=2, label='Standard Normal PDF')
    plt.axvline(z_score, color='red', linestyle='dashed', lw=2, label=f'Z-score = {z_score:.2f}')

    plt.xlabel('Z-Score')
    plt.ylabel('Density')
    plt.title('Comparison with Standard Normal Distribution')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    return z_score

def hypothesis_testing(sample, pop_mean, pop_std, alpha=0.05):
    """Perform hypothesis testing using Z-statistics."""
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    standard_error = pop_std / np.sqrt(sample_size)
    z_stat = (sample_mean - pop_mean) / standard_error

    # Calculate p-value
    p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    # Decision
    reject_null = p_value < alpha

    print(f"Z-Statistic: {z_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")
    print("Reject Null Hypothesis" if reject_null else "Fail to Reject Null Hypothesis")

    return z_stat, p_value, reject_null

def central_limit_theorem(sample_size, num_samples):
    """Demonstrate the Central Limit Theorem using a uniform distribution."""
    population = np.random.uniform(0, 10, 100000)  # Non-normal distribution
    sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

    # Plot sample means histogram
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='purple', edgecolor='black')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.title(f'Central Limit Theorem Demonstration (n={sample_size})')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    return sample_means

def confidence_interval(data, confidence=0.95):
    """Calculate and plot the confidence interval for a dataset."""
    sample_mean = np.mean(data)
    sample_size = len(data)
    standard_error = np.std(data, ddof=1) / np.sqrt(sample_size)
    margin_of_error = stats.norm.ppf((1 + confidence) / 2) * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"{confidence*100}% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

    # Plot confidence interval
    plt.figure(figsize=(6, 4))
    plt.errorbar(1, sample_mean, yerr=margin_of_error, fmt='o', color='red', capsize=5, label='Confidence Interval')
    plt.axhline(y=lower_bound, color='blue', linestyle='dashed', label='Lower Bound')
    plt.axhline(y=upper_bound, color='blue', linestyle='dashed', label='Upper Bound')
    plt.xlabel('Sample')
    plt.ylabel('Mean Value')
    plt.title(f'{confidence*100}% Confidence Interval')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    return lower_bound, upper_bound

# Parameters
mu, sigma = 0, 1  # Mean and standard deviation
data = simulate_normal(mu, sigma)
calculate_z_scores(data)

# Example Z-score calculation
random_data_point = np.random.choice(data)
z_score = calculate_z_score(random_data_point, data)
print(f'Z-score of {random_data_point}: {z_score:.2f}')

# CLT Demonstration with Non-Normal Distribution
sample_size = 30  # Number of elements in each sample
num_samples = 1000  # Number of samples
distributed_means = central_limit_theorem(sample_size, num_samples)

# Hypothesis Testing Example
sample = np.random.normal(mu, sigma, 50)  # Generate a sample
hypothesis_testing(sample, pop_mean=0, pop_std=1, alpha=0.05)

# Confidence Interval Example
confidence_interval(sample, confidence=0.95)
```

```python
#Q(17)Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean?
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def generate_normal_data(mean=0, std_dev=1, size=100):
    """Generate random data from a normal distribution."""
    return np.random.normal(mean, std_dev, size)

def confidence_interval(data, confidence=0.95):
    """Calculate and plot the confidence interval for the mean of a dataset."""
    sample_mean = np.mean(data)
    sample_size = len(data)
    standard_error = np.std(data, ddof=1) / np.sqrt(sample_size)
    margin_of_error = stats.norm.ppf((1 + confidence) / 2) * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"{confidence*100}% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")

    # Plot confidence interval
    plt.figure(figsize=(6, 4))
    plt.errorbar(1, sample_mean, yerr=margin_of_error, fmt='o', color='red', capsize=5, label='Confidence Interval')
    plt.axhline(y=lower_bound, color='blue', linestyle='dashed', label='Lower Bound')
    plt.axhline(y=upper_bound, color='blue', linestyle='dashed', label='Upper Bound')
    plt.xlabel('Sample')
    plt.ylabel('Mean Value')
    plt.title(f'{confidence*100}% Confidence Interval for Mean')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    return lower_bound, upper_bound

# Generate data and calculate confidence interval
data = generate_normal_data(mean=50, std_dev=10, size=100)
confidence_interval(data, confidence=0.95)
```

```python
#Q(18)Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution?
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_normal_pdf(mean=0, std_dev=1):
    """Calculate and visualize the probability density function (PDF) of a normal distribution."""
    # Generate x values around the mean
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

    # Compute the normal PDF values
    pdf = stats.norm.pdf(x, mean, std_dev)

    # Plot the PDF
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, 'r-', lw=2, label=f'N({mean}, {std_dev}Â²)')
    plt.fill_between(x, pdf, alpha=0.2, color='red')  # Shaded region under curve

    # Labels and title
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.title(f'Normal Distribution PDF (Î¼={mean}, Ïƒ={std_dev})')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)

    # Show the plot
    plt.show()

# Example Usage
plot_normal_pdf(mean=0, std_dev=1)
```

```python
#Q(19)Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution?
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def poisson_cdf(lam=5, max_k=20):
    """Calculate and visualize the CDF of a Poisson distribution."""
    # Generate discrete values for k (number of events)
    k_values = np.arange(0, max_k + 1)

    # Compute the Poisson CDF
    cdf_values = stats.poisson.cdf(k_values, lam)

    # Plot the CDF
    plt.figure(figsize=(8, 5))
    plt.step(k_values, cdf_values, where='mid', color='blue', lw=2, label=f'Poisson CDF (Î»={lam})')
    plt.scatter(k_values, cdf_values, color='red', zorder=3)

    # Labels and title
    plt.xlabel('Number of Events (k)')
    plt.ylabel('Cumulative Probability')
    plt.title(f'Poisson Distribution CDF (Î»={lam})')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)

    # Show the plot
    plt.show()

    return k_values, cdf_values

# Example Usage
poisson_cdf(lam=5, max_k=20)
```

```python
#Q(20)Simulate a random variable using a continuous uniform distribution and calculate its expected value?
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def simulate_uniform(a=0, b=10, size=10000):
    """Simulate a continuous uniform random variable and calculate expected value."""
    # Generate uniform random samples
    samples = np.random.uniform(a, b, size)

    # Calculate expected value (mean of uniform distribution)
    expected_value = (a + b) / 2

    # Plot histogram
    plt.figure(figsize=(8, 5))
    plt.hist(samples, bins=30, density=True, alpha=0.7, color='blue', edgecolor='black')

    # Plot theoretical PDF
    x = np.linspace(a, b, 1000)
    pdf = stats.uniform.pdf(x, a, b-a)
    plt.plot(x, pdf, 'r-', lw=2, label='Uniform PDF')

    # Labels and title
    plt.xlabel('Value')
    plt.ylabel('Density')
    plt.title(f'Continuous Uniform Distribution U({a}, {b})')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)

    # Show plot
    plt.show()

    print(f"Expected Value (E[X]) = {expected_value}")

    return samples, expected_value

# Example Usage
simulate_uniform(a=0, b=10, size=10000)
```

```python
#Q(21)Write a Python program to compare the standard deviations of two datasets and visualize the difference?
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def compare_standard_deviations(size=1000, mean1=50, std1=10, mean2=50, std2=20):
    """Generate two datasets with different standard deviations and visualize them."""

    # Generate random data from normal distributions
    data1 = np.random.normal(mean1, std1, size)
    data2 = np.random.normal(mean2, std2, size)

    # Compute standard deviations
    std_dev1 = np.std(data1, ddof=1)
    std_dev2 = np.std(data2, ddof=1)

    # Plot histograms
    plt.figure(figsize=(10, 5))
    plt.hist(data1, bins=30, alpha=0.6, color='blue', label=f'STD1={std_dev1:.2f}')
    plt.hist(data2, bins=30, alpha=0.6, color='red', label=f'STD2={std_dev2:.2f}')

    # Titles and labels
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Comparison of Distributions with Different Standard Deviations')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)

    # Show histogram
    plt.show()

    # Boxplot for visualization
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=[data1, data2], palette=['blue', 'red'])
    plt.xticks([0, 1], ['Dataset 1', 'Dataset 2'])
    plt.ylabel('Values')
    plt.title('Boxplot Comparison of Standard Deviations')
    plt.grid(True, linestyle='--', alpha=0.5)

    # Show boxplot
    plt.show()

    # Print results
    print(f"Dataset 1 Standard Deviation: {std_dev1:.2f}")
    print(f"Dataset 2 Standard Deviation: {std_dev2:.2f}")

# Example Usage
compare_standard_deviations()
```

```python
#Q(22)Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution?
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def compute_range_and_iqr(size=1000, mean=50, std=15):
    """Generate a normal dataset, compute range & IQR, and visualize."""

    # Generate normal distribution data
    data = np.random.normal(mean, std, size)

    # Compute statistics
    data_range = np.max(data) - np.min(data)
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1  # Interquartile range

    # Print results
    print(f"Range: {data_range:.2f}")
    print(f"Interquartile Range (IQR): {IQR:.2f}")
    print(f"Q1 (25th percentile): {Q1:.2f}")
    print(f"Q3 (75th percentile): {Q3:.2f}")

    # Boxplot for visualization
    plt.figure(figsize=(7, 5))
    sns.boxplot(data, color='skyblue')
    plt.title("Boxplot of Normal Distribution")
    plt.xlabel("Values")
    plt.grid(True, linestyle='--', alpha=0.7)

    # Show boxplot
    plt.show()

    return data_range, IQR

# Example Usage
compute_range_and_iqr()
```

```python
#Q(23)Implement Z-score normalization on a dataset and visualize its transformation?
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def z_score_normalization(size=1000, mean=50, std=15):
    """Generate a dataset, normalize using Z-score, and visualize transformation."""

    # Generate dataset from a normal distribution
    data = np.random.normal(mean, std, size)

    # Compute Z-score normalization
    normalized_data = (data - np.mean(data)) / np.std(data)

    # Plot original and normalized distributions
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    # Original Data Histogram
    sns.histplot(data, bins=30, kde=True, color='blue', ax=axes[0])
    axes[0].set_title("Original Data Distribution")
    axes[0].set_xlabel("Value")
    axes[0].set_ylabel("Frequency")

    # Normalized Data Histogram
    sns.histplot(normalized_data, bins=30, kde=True, color='red', ax=axes[1])
    axes[1].set_title("Z-Score Normalized Distribution")
    axes[1].set_xlabel("Z-Score")
    axes[1].set_ylabel("Frequency")

    plt.tight_layout()
    plt.show()

    # Print mean and standard deviation after normalization
    print(f"Original Mean: {np.mean(data):.2f}, Original STD: {np.std(data):.2f}")
    print(f"Normalized Mean: {np.mean(normalized_data):.2f}, Normalized STD: {np.std(normalized_data):.2f}")

    return normalized_data

# Example Usage
z_score_normalization()
```

```python
#(Q24)Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.?
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis

def compute_skewness_kurtosis(size=1000, mean=0, std=1):
    """Generate a normal dataset, compute skewness & kurtosis, and visualize."""

    # Generate normal distribution data
    data = np.random.normal(mean, std, size)

    # Compute skewness & kurtosis
    skewness = skew(data)
    kurt = kurtosis(data)  # Excess kurtosis (subtracts 3)

    # Print results
    print(f"Skewness: {skewness:.4f} (â‰ˆ0 for normal distribution)")
    print(f"Kurtosis (Excess): {kurt:.4f} (â‰ˆ0 for normal distribution)")

    # Histogram + KDE Plot
    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, kde=True, color='blue', edgecolor='black')
    plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label='Mean')

    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.title("Histogram of Normal Distribution with KDE")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)

    # Show plot
    plt.show()

    return skewness, kurt

# Example Usage
compute_skewness_kurtosis()
```
